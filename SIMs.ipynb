```python
# Imports
import calendar
import random as rnd
from datetime import date, timedelta
import lorem
import csv
import datetime
from dateutil.relativedelta import relativedelta
```

```python
def generate_mobile():
    number = "07"
    
    for i in range(9):
        number += str(rnd.randint(0,9))
    
    return number

def generate_landline():
    area_codes = ['01934', '01179', '01275', '01823']
    
    number = rnd.choice(area_codes)
    
    for i in range(6):
        number += str(rnd.randint(0,9))
        
    return number

def generate_dob(min_age=19, max_age=50):
    """Generates a rnd date of birth within the specified age range."""

    today = date.today()
    max_years = today.year - min_age
    min_years = today.year - max_age

    # Generate a rnd year within the age range
    rnd_year = rnd.randint(min_years, max_years)

    # Generate a rnd month (1-12)
    rnd_month = rnd.randint(1, 12)

    # Calculate the maximum number of days in the chosen month
    _, days_in_month = calendar.monthrange(rnd_year, rnd_month)

    # Generate a rnd day within the month
    rnd_day = rnd.randint(1, days_in_month)

    # Create a date object with the generated values
    rnd_dob = date(rnd_year, rnd_month, rnd_day)

    # Format the date as dd/mm/yyyy
    dob_string = rnd_dob.strftime("%d/%m/%Y")

    return dob_string


def generate_child_dob(min_age=1, max_age=17):
    """Generates a rnd date of birth within the specified age range."""

    today = date.today()
    max_years = today.year - min_age
    min_years = today.year - max_age

    # Generate a rnd year within the age range
    rnd_year = rnd.randint(min_years, max_years)

    # Generate a rnd month (1-12)
    rnd_month = rnd.randint(1, 12)

    # Calculate the maximum number of days in the chosen month
    _, days_in_month = calendar.monthrange(rnd_year, rnd_month)

    # Generate a rnd day within the month
    rnd_day = rnd.randint(1, days_in_month)

    # Create a date object with the generated values
    rnd_dob = date(rnd_year, rnd_month, rnd_day)

    # Format the date as dd/mm/yyyy
    dob_string = rnd_dob.strftime("%d/%m/%Y")

    return dob_string

def generate_name():
    first_names = ['Liam', 'Ava', 'Noah', 'Olivia', 'Elijah', 'Sophia', 'William', 'Isabella', 
                   'James', 'Charlotte', 'Lucas', 'Mia', 'Mason', 'Amelia', 'Ethan', 'Evelyn', 
                   'Benjamin', 'Abigail', 'Henry', 'Harper', 'Alexander', 'Audrey', 'Michael', 
                   'Sofia', 'Daniel', 'Elizabeth', 'Aiden', 'Eleanor', 'Matthew', 'Lilah', 'Logan', 
                   'Madison', 'Jackson', 'Scarlett', 'Sebastian', 'Penelope', 'Owen', 'Riley', 'Samuel', 
                   'Savannah', 'Gabriel', 'Chloe', 'Carter', 'Stella', 'Levi', 'Victoria', 'Isaac', 
                   'Nora', 'Julian', 'Luna']
    
    surnames = ['Mitchell', 'Allen', 'Nguyen', 'King', 'Anderson', 'Wilson', 'Jackson', 'Young', 
                'Thompson', 'Sanchez', 'Lewis', 'Taylor', 'Robinson', 'Parker', 'Lee', 'Smith', 
                'Davis', 'Wright', 'Martin', 'Phillips', 'Garcia', 'Miller', 'White', 'Lopez', 
                'Walker', 'Garcia', 'Thomas', 'Brown', 'Harris', 'Jones', 'Robinson', 'Hernandez', 
                'Young', 'Taylor', 'King', 'Lopez', 'Williams', 'Moore', 'Clark', 'Gonzalez', 
                'Lewis', 'Wright', 'Walker', 'Robinson', 'Hernandez', 'Johnson', 'Rodriguez', 
                'Diaz', 'Allen', 'Clark']
    
    first_name = rnd.choice(first_names)
    surname = rnd.choice(surnames)
    
    return f'{first_name} {surname}'
```

```python
def Aggrieved_details(mobile=False, landline=False, other=False):
    
    result = {}
    
    if mobile == True:
        mobile_number = generate_mobile()
        result['Aggrieved mobile'] = mobile_number
        
    if landline == True:
        landline_number = generate_landline()
        result['Aggrieved landline'] = landline_number
                
    if other == True:
        other_number = generate_mobile()
        result['Aggrieved other'] = other_number
        
    result['Full name of aggrieved'] = generate_name()
    result['Full name of offender'] = generate_name()
    result['DOB of offender'] = generate_dob()
    
    return result
    
```

```python
def Offender_details():
    
    result = {}
    
    result['Full name of offender'] = generate_name()
    result['DOB of offender'] = generate_dob()
    
    return result
```

```python
def Occupant_details(mobile=False, landline=False, other=False):
    
    result = {}
    
    if mobile == True:
        mobile_number = generate_mobile()
        result['Occupant mobile'] = mobile_number
        
    if landline == True:
        landline_number = generate_landline()
        result['Occupant landline'] = landline_number
                
    if other == True:
        other_number = generate_mobile()
        result['Occupant other'] = other_number
        
    result['Full name of occupant'] = generate_name()
    result['DOB of occupant'] = generate_dob()
    
    return result
```

```python
def Generate_children(number_kids):
    
    result = {}
    
    for i in range(number_kids):
        result['Child ' + str(i + 1) + ' - name'] = generate_name()
        result['Child ' + str(i + 1) + ' - DOB'] = generate_child_dob()
        
    return result
```

```python
# Generate marker text
def Marker_text(guidance = "other"):
    
    marker_text = ""

    if guidance == "Court Order":
        marker_text = "Parties are subject to a Court Order. PNC should be checked to verify the full terms and expiry date of the Order."

    elif guidance == "DVPN":
        marker_text = "A Domestic Violence Protection Notice (DVPN) has been issued to the OFFD pending the issue of a Domestic Violence Protection Order (DVPO). Please refer to Niche for full details."
        
    elif guidance == "DVPO":
        marker_text = "A Domestic Violence Protection Order has been issued by the courts. Please refer to PNC for full condition details"
        
    elif guidance == "Vulnerable Child":
        marker_text = "Attending Officers must check Child’s / Children’s welfare and update LSU via Niche Workflow."
        
    elif guidance == "other":
        for i in range (3):
            marker_text = marker_text + lorem.sentence() + " "
                
    return marker_text
```

```python
def Email_notifications(email_me=False, email_other=False):

    email_notifications = {}

    if email_me == True:
        email_notifications['Email'] = "Yes"
    else:
        email_notifications['Email'] = "No"

    if email_other == True:
        email_notifications['Email other?'] = "Yes"
        email_notifications['Collar number email'] = rnd.randint(1000,4999)
    else:
        email_notifications['Email other'] = "No"

    return email_notifications
```

```python
def Rand_Other():
    """Create a ASB Victim, DA or Risk of Harm marker.
    """
        
    result = {}
    
    # Marker type
    marker_type_q = "Type of marker"
    marker_type_a = rnd.choice(["ASB Victim", "Domestic Abuse", "Risk of Harm"])
    result[marker_type_q] = marker_type_a
    
    # Risk Question
    risk_q = risk_questions['Other']
    risk_a = rnd.choice(levels_of_risk)    
    result[risk_q] = risk_a
    
    # Time active
    weed_q = "Weed"
    if marker_type_a == "Domestic Abuse" and risk_a == "High Risk":
        weed_a = "12 months"
    else:
        weed_a = rnd.choice(weed_dates)    
    result[weed_q] = weed_a
    
    # Harm question
    harm_q = 'Harm'
    harm_a =  rnd.choice(type_of_harm[harm_q])    
    result[harm_q] = harm_a
    
    # Victim Contact Details     
    random_list = rnd.choices([True, False], k=3)    
    contacts = Aggrieved_details(random_list[0], random_list[1], random_list[2])
    result.update(contacts)
    
    # Offender details
    
    offender = Offender_details()
    result.update(offender)
    
    # Marker text
    marker_text_q = "Marker text"
    marker_text_a = Marker_text()    
    result[marker_text_q] = marker_text_a
    
    rand_email = rnd.choices([True, False], k=2)
    email_note = Email_notifications(rand_email[0], rand_email[1])
    
    result.update(email_note)
    
    return result
```

```python
csv_path = "Test Addresses copy.csv"

with open(csv_path, 'r', newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    data = list(reader)  

marker_types = ['Other', 'Other', 'Other', 'Court', 'DVPN', 'DVPO', 'Harrass', 'MH', 'Child', 
                'Officer', 'Vuln Person']

result_list = []

for i in range(200):
    
    marker_choice = rnd.choice(marker_types)
    
    match marker_choice:
        case 'Other':
            details = Rand_Other()
        case 'Court':
            details = Rand_Court_Order()
        case 'DVPN':
            details = Rand_DVPN()
        case 'DVPO':
            details = Rand_DVPO()
        case 'Harrass':
            details = Rand_Harrassment()
        case 'MH':
            details = Rand_Mental_Health()
        case 'Child':
            details = Rand_Vulnerable_Child()
        case 'Officer':
            details = Rand_Officer_Safety()
        case 'Vuln Person':
            details = Rand_Vulnerable_Persons()

    rw = data[i]
    
    for key, value in details.items():
        if key in rw:
            rw[key] = value            

        # Copy address into right field, with html elements

    address = f"<p>{rw['line_1']}<br>{rw['line_2']}<br>{rw['town']}<br>{rw['postcode']}<br>United Kingdom</p>"
    rw["Address for information marker"] = address
    rw["UPRN"] = rw["UPRN In"]
    
    result_list.append(rw)
        
# for d in result_list:
#     print(d)    
```
